```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import re

film_ori = pd.read_csv("../data/film.csv")

film = film_ori.copy()

film.head(5)
```

```python
#a tomar por saco la columna last_update
film = film.drop("last_update", axis=1)
```

```python
#me cargo tambien la columna original_lenguage_id, eesa columna está tan vacia como el corazón de ella
film = film.drop("original_language_id", axis=1)
```

```python
film.head(50)
```

```python
film["language_id"].unique()
```

```python
#la columna language_id solo tiene un valor, a tomar por el saco
film = film.drop("language_id", axis=1)
```

```python
#añadimos la columna language y ponemos english porque eran todas en inglés
film["language"] = "english"
```

```python
film.head(5)
```

```python
#pasamos a minuscula tanto title como description como special_features para seguir unificando formatos de forma estetica
film["title"] = film["title"].str.lower()
film["description"] = film["description"].str.lower()
film["special_features"] = film["special_features"].str.lower()
```

```python
#damos por terminada la limpieza de este archivo
film.to_csv("../data/filmclean.csv", index=False)
```
